```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns 

file_name = 'variant_04'
df = pd.read_csv(file_name + '.csv')
df.info(verbose=True)
# print(df)


# dfc.to_csv('clear.csv')
```

 ### Блок удаления повторений

 Метод ___drop_duplicates___ позволяет удалить все повтореня, оставляя только первое вхождение

```python
dfc = df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
dfc.info()
```

### Блок удаления "пустых данных"

Метод ___dropna___ позволяет без дополнительного анализа удалить строки с пустыми ячейками

```python
dfc = dfc.dropna()
dfc.info()
```

 ### Блок удаления аномальных данных

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.displot(dfc['MEDV'], bins=30)
plt.show()

dfc = dfc.reset_index(drop=True)
print('++++++++------------------')
dfc.info()
print('++++++++++-------------')
for i in range(0,len(dfc['MEDV'])):
    if dfc['MEDV'][i] < 0:
        dfc = dfc.drop([i])
dfc.info()
dfc.to_csv('clear.csv')
for i in range(0,len(dfc.columns)):
    sns.set(rc={'figure.figsize':(10,10)})
    sns.displot(dfc[dfc.columns[i]], bins=30)
    plt.show()

```
